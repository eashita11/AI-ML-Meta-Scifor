Que 3- What is Anova test. how we can calculate behind the scene ?

**Ans**=
**Analysis of Variance(ANOVA)** is used to compare the differences in means among more than two groups. It examines the variation in the data and identifying where this variation is located. Specifically, ANOVA compares the amount of variation between groups with the amount of variation within groups.

Behind the scenes, it works in the following way:
We first calculate the mean of each group, and then the grand mean. Then we calculate the variations within and between the groups. Then calculate the mean squared variance between and within the groups. Then calculate F statistics and corresponding p value. 

**F-value** : The sample means of betweeen groups /sample means of within groups.

**p values** is the value we get after reffering to the z table through the z value. Its the probabilty value.

Que 2-  Suppose a fitness center manager anticipates the attendance for a particular yoga class throughout the week. Based on historical data and other factors, they estimate the expected attendance in numbers:

### Expected Attendance:

Monday: 40

Tuesday: 45

Wednesday: 38

Thursday: 50

Friday: 55

Saturday: 60

Sunday: 65

However, after the week concludes, the actual attendance is recorded:



### Observed (Actual) Attendance:



Monday: 38

Tuesday: 43

Wednesday: 40

Thursday: 48

Friday: 52

Saturday: 58

Sunday: 62

Ho: There is no significant difference between the expected and observed attendance for the yoga class throughout the week.

Ha: There is a significant difference between the expected and observed attendance for the yoga class throughout the week.

#### Use chi-square test *Statistical Tools*

Sol=

```python
import numpy as np

# Expected and Observed values
expected = [40, 45, 38, 50, 55, 60, 65]
observed = [38, 43, 40, 48, 52, 58, 62]

# Calculate the Chi-Square statistic for each pair
chi_ind = [
    np.square(observed[i] - expected[i]) / expected[i] 
    for i in range(len(expected))
]

# Sum the Chi-Square statistics to get the total Chi-Square value
chi_square_stat = np.sum(chi_ind)

print("Our Chi-Square test statistic score ==> ", chi_square_stat)
```

Check the chi square table, and we get that the critical score is greater than the test score, which means we cannot reject the null hypothesis, which means that the prediction, there is no significant difference between the expected and observed attendance for the yoga class throughout the week, is correct.

Que3 -> Calculate anova test





    | Student_ID | Semester | Total_Marks |
    |------------|----------|-------------|
    | 1          | Fall     | 85          |
    | 2          | Fall     | 78          |
    | 3          | Fall     | 92          |
    | 4          | Spring   | 88          |
    | 5          | Spring   | 75          |
    | 6          | Spring   | 82          |
    | 7          | Summer   | 90          |
    | 8          | Summer   | 86          |
    | 9          | Summer   | 79          |    


Null Hypothesis (Ho): The average marks of students in each semester are approximately equal.

Alternative Hypothesis (Ha): The average marks of students in at least one semester are different.

Sol=

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

data = {
    'Student_ID': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Semester': ['Fall', 'Fall', 'Fall', 'Spring', 'Spring', 'Spring', 'Summer', 'Summer', 'Summer'],
    'Total_Marks': [85, 78, 92, 88, 75, 82, 90, 86, 79]
}

df = pd.DataFrame(data)
mod = smf.ols('Total_Marks ~ Semester', data=df).fit()

anova_table = sm.stats.anova_lm(mod, typ=1)
anova_table
```

The ANOVA test shows that there is no significant difference in the average marks of students across Fall, Spring, and Summer semesters (since p-value > 0.05).
