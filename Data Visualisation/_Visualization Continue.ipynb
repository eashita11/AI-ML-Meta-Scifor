```python
sal=[15,22,24,25,30,23,34,43,19,20,52,21]
sal.sort()
sal
```

# Please see the below Example




[15,19,20,21,22,23,24,25,30,34,43,52]




n=total elements in the above (12 items)



--> 25%=(25/100) * (n+1)=3.25(data at 3.25 Position is the 25% percentile .lets say around 20.50)




--> 50%=(50/100) * (n+1)=6.5(data at this position)




--> 75%=(75/100) * (n+1)=9.75(data at this position)

# Outliers

```python
import pandas as pd
ser=pd.Series(sal)
ser
```

```python
#1st Quantile
q1=ser.quantile(0.25)
print('Q1',q1)
#3rd Quantile
q3=ser.quantile(0.75)
print('Q3',q3)
#Find the IQR (inter quantile range) to identify outliers
iqr=q3-q1
print('IQR',iqr)
```

# Outliers Detection Formula
higher side==> Q3 + (1.5 * IQR)
lower side ==> Q1 - (1.5 * IQR)

```python
upside=q3 + (1.5 * iqr)
print(upside)
```

```python
#Anything above 43.373 is outlier
```

```python
lower_side=q1-(1.5 * iqr)
print(lower_side)
```

```python
#Anything below 5.375 is outlier
```
